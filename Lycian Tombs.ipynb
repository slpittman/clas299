```scala
import scala.io.Source
val tombset = "http://shot.holycross.edu/courses/ada/S20/data/lycianNecropoleis.cex"
```

```scala
val lines = Source.fromURL(tombset).getLines.toVector
```

```scala
val separated = lines.tail.map(ln => ln.split("#").toVector)
```

```scala
val tombcount = separated.map(columns => columns(2))
```

```scala
val integercount = tombcount.map(t => t.toInt)
```

```scala
integercount.sum
```
